# Task 1.a
Compute the Fourier transform for the function:  $f(x) = x e^{-\alpha |x|}, \alpha > 0$   

$$F(\omega)=\int_{-\infty}^{+\infty} f(x) e^{-j \omega x} d x=\int_{-\infty}^{0} x e^{\alpha x} e^{-j \omega x} d x+\int_{0}^{+\infty} x e^{-\alpha x} e^{-j \omega x} d x = I_{1} + I_{2}$$ 


$$
I_{1}=\int_{-\infty}^{0} x e^{(\alpha-j \omega) x} d x 
=\frac{1}{\alpha-j \omega} \int_{-\infty}^{0} x d e^{(\alpha-j \omega) x} =\\
=\frac{1}{\alpha-j \omega} \left[xe^{(\alpha-j \omega) x}|_{-\infty} ^{0}-\frac{1}{\alpha-j \omega} \int_{-\infty}^{0} e^{(\alpha-j \omega) x} d ((\alpha -j \omega)x) \right]=\\
=\frac{1}{\alpha-j\omega}\left[0-\frac{1}{\alpha-j \omega}x e^{(\alpha-j \omega) x}|_{-\infty} ^{0}\right] = \\
=\frac{1}{\alpha-j \omega}\left[-\frac{1}{\alpha-j \omega}(1-0)\right]=\frac{-1}{(\alpha-j \omega)^{2}}
$$

Note: $$\left.x e^{(\alpha-j \omega) x}\right|_{-\infty} ^{0}=0 - \lim _{x \rightarrow-\infty} x e^{(\alpha-j \omega) x}=\\
= - \lim _{x \rightarrow-\infty} \frac {x}{e^{-(\alpha-j \omega)x}}= [\frac{\infty}{\infty}] = by~L'Hopital's~rule = \\ =- \lim _{x \rightarrow-\infty} \frac {1}{-(\alpha-j \omega)e^{-(\alpha-j \omega) x}} = 0$$

$$
I_{2}=\int_{0}^{+\infty} x e^{-(\alpha+j \omega) x} d x= 
-\frac{1}{\alpha+j \omega} \int_{0}^{+\infty} x d e^{-(\alpha+j \omega) x}= \\ =
-\frac{1}{\alpha+j \omega}\left[\left.x e^{-(\alpha+j\omega)}|_{0} ^{+\infty}\right. 
+\frac{1}{\alpha+j \omega} \int_{0}^{+\infty} e^{-(\alpha+j \omega) x} d(-(\alpha+j \omega) x)\right]= \\
=-\frac{1}{(\alpha+j \omega)^{2}}(0-1)=\frac{1}{(\alpha+j \omega)^{2}}
$$

Note: $$ \left.x e^{-(\alpha+j \omega) x}\right|_{0} ^{+\infty} = \lim _{x \rightarrow+\infty} x e^{-(\alpha+j \omega) x} - 0 =\\= \lim _{x \rightarrow+\infty} \frac {x}{e^{(\alpha+j \omega)x}}= [\frac{\infty}{\infty}] = by~L'Hopital's~rule = \\ =\lim _{x \rightarrow+\infty} \frac {1}{(\alpha+j \omega)e^{(\alpha+j \omega) x}}= 0$$ 

Finally: $$I = I_{2} + I_{1} = \frac{1}{(\alpha+j \omega)^{2}} + \frac{-1}{(\alpha-j \omega)^{2}}  = \\=\frac{\alpha^{2}-2j\omega \alpha - \omega^{2}-\alpha^{2}-2j\omega +\omega^{2}}{(\alpha+j\omega) (\alpha + j\omega)(\alpha -j \omega)(\alpha - j \omega)} =  \frac{-4j\omega\alpha}{(\alpha^{2}+\omega^{2})^{2}}$$

**Answer:** $$F(\omega) =  \frac{-4j\omega\alpha}{(\alpha^{2}+\omega^{2})^{2}}$$

# Task 1.b
Compute the Fourier transform for the function:  $f(x) = e^{-a^{2}x^{2}}cos(bx)$

Here I will use the shifting property of FT. For this reason I will start my calculations from computing the FT of function: $f(x) = e^{-a^{2}x^{2}}$ and then will apply shift to the result.

$$F_{1}(\omega)=\int_{-\infty}^{+\infty} e^{-a^{2} x^{2}} e^{-j \omega x}dx= \\ =\int_{-\infty}^{+\infty} e^{-(a^{2} x^{2} + j \omega x -\frac{\omega^{2}}{4a^{2}}+\frac{\omega^{2}}{4a^{2}})} d x =\\
=e^{-\frac{\omega^{2}}{4a^{2}}} \int_{-\infty}^{+\infty} e^{-\left(|a| x+\frac{j \omega}{2|a|}\right)^{2}} d x=\\
=e^{-\frac{\omega^{2}}{4a^{2}}} \frac{1}{|a|} \int_{-\infty}^{+\infty} e^{-\left(|a| x+\frac{j\omega}{2|a|}\right)^{2}} d\left(|a| x+\frac{j \omega}{2|a|}\right)
=\frac{\sqrt{\pi}}{|a|} e^{-\frac{\omega^{2}}{4a^{2}}}$$

Applying shifting property:

$$F(\omega)=\int_{-\infty}^{+\infty} e^{-\alpha^{2} x^{2}} \cos (b x) e^{-j \omega x} d x=\frac{1}{2} \int_{-\infty}^{+\infty}e^{-a^{2} x^{2}}\left(e^{j b x}+e^{-j b x}\right) e^{-j \omega x} d x = \\
=\frac{1}{2}\left(\int_{-\infty}^{+\infty} e^{-a^{2} x^{2}} e^{-j(\omega-b) x} d x+\int_{-\infty}^{+\infty} e^{-a^{2} x^{2}} e^{-j(\omega+b) x} d x\right)= \\
=\frac{1}{2}\left(\frac{\sqrt{\pi}}{|a|} e^{-\frac{(\omega-b)^{2}}{4 a^{2}}}+\frac{\sqrt{\pi}}{|a|} e^{-\frac{(\omega+b)^{2}}{4 a^{2}}}\right)
=\frac{\sqrt{\pi}}{2|a|}\left(e^{-\frac{(\omega-b)^{2}}{4 a^{2}}}+e^{-\frac{(\omega+ b)^{2}}{4 a^{2}}}\right)$$

**Answer:** $$F(\omega) = \frac{\sqrt{\pi}}{2|a|}\left(e^{-\frac{(\omega-b)^{2}}{4 a^{2}}}+e^{-\frac{(\omega+ b)^{2}}{4 a^{2}}}\right)$$

# Task 2
Compute the spectrum of the cosine-filled sinc function:

```python
#plot picture from HW file
from IPython.display import Image
display(Image(filename='sinc.png', width = 400))
```

First of all, I defined the cosine-filled sinc function: $f(x)=\frac{\sin (\pi x)}{\pi x} \cos \left(\omega_{0} x+\varphi_{0}\right)$. I write $\omega_{0}, \varphi_{0}$ because I want to derive the representation of cosine-filled sinc function in the frequency domain in general case.  
By the way, it is obvious that for our case $\varphi_{0} = -\frac{\pi}{2}$ (because cosine is delayed ($f(0) = 0$)) and $\omega_{0} = 3\pi$  
Proof: $f(x)=\frac{\sin (\pi x)}{\pi x} \cos (\omega_{0} x-\frac{\pi}{2}) = 0$,  
then   
$$\left[ 
      \begin{gathered} 
        sinc(x) = 0 \\ 
        cos(\omega_{0}x - \frac{\pi}{2}) = 0 \\ 
      \end{gathered} 
\right.$$
from the second equation we get  
$$\omega_{0}x-\frac{\pi}{2} = \frac{\pi}{2} + \pi k, k \in N \\
\omega_{0}x = \pi + \pi k$$
Then according to the graph above I can say that when $x = \frac{1}{3}$ cosine-filled function equals to zero, so the first root (when $k = 0$) of the equation above can be find in this way:
$$\frac{\omega_{0}}{3} = \pi \\
\omega_{0} = 3\pi$$ 

$$
F(\omega)=\int_{-\infty}^{+\infty} sinc(x) \cos \left(\omega_{0} x+\varphi_{0}\right) e^{-j \omega x} d x= \\ =\int_{-\infty}^{+\infty}sinc(x)\left(\frac{e^{j\left(\omega_{0} x+\varphi_{0}\right)}+e^{-j\left(\omega_{0} x+\varphi_{0}\right)}}{2}\right) \cdot e^{-j \omega x} d x= \\
=\frac{1}{2} \left(\int_{-\infty}^{+\infty} e^{j \varphi_{0}}sinc(x) e^{-j\left(\omega-\omega_{0}\right) x} d x+\int_{-\infty}^{+\infty} e^{-j \varphi_{0}} sinc(x) e^{-j\left(\omega+\omega_{0}\right) x} d x\right)=I_{1}+I_{2}
$$

$$
I_{1}=\frac{1}{2} e^{j \varphi_{0}}\int_{-\infty}^{+\infty}sinc(x) e^{-j\left(\omega-\omega_{0}\right) x} d x=\left\{\begin{array}{c}
\frac{1}{2} e^{j \varphi_{0}},\left|\omega-\omega_{0}\right| \leqslant \pi \\
0, \text {otherwise} \hfill
\end{array}\right. \\
I_{2}=\frac{1}{2} e^{-j \varphi_{0}}\int_{-\infty}^{+\infty}sinc(x) e^{-j\left(\omega+\omega_{0}\right) x} d x=\left\{\begin{array}{c}
\frac{1}{2} e^{-j \varphi_{0}},\left|\omega+\omega_{0}\right| \leqslant \pi \\
0, \text {otherwise} \hfill
\end{array}\right. 
$$
For our case $\omega_{0} = 3\pi, \varphi_{0} = -\frac{\pi}{2}$ we get:
$$
I_{1}=\left\{\begin{array}{c}
\frac{j}{2},2\pi \leqslant \omega \leqslant 4\pi \\
0, \text {otherwise} \hfill
\end{array}\right. \\
$$
$$
I_{2}=\left\{\begin{array}{c}
-\frac{j}{2},-4\pi \leqslant \omega \leqslant -2\pi \\
0, \text {otherwise} \hfill
\end{array}\right. \\
$$



Finally:
$$
I=\left\{\begin{array}{c}
-\frac{j}{2},-4\pi \leqslant \omega \leqslant -2\pi \\
\frac{j}{2},2\pi \leqslant \omega \leqslant 4\pi \\
0, \text {otherwise} \hfill
\end{array}\right. \\
$$


Now let's plot the graph. Firstly, we need to find the magnitude of the signal in frequency domain. In our case it is obvious that $A = \frac{1}{2}$ (by definition $A(\omega) = \sqrt{F(\omega)F^{*}(\omega)}$ )

```python
x = np.linspace(-13, 13, 260)
y = []
for point in x:
    if (-4 * np.pi <= point <= -2 * np.pi):
        y.append(0.5)
    elif (2 * np.pi <= point <= 4 * np.pi): # of course, I can fill y 
                                            # list in one if clause 
        y.append(0.5)                       #but I did it in this 
                                            #way for the better understanding
    else:
        y.append(0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(x, y)
ax.grid()
ax.set_title('F(w)');
```

We recieved the rectangular window in frequency domain shifted by $\omega_{0} = 3\pi$ with two times decreased magnitude. 

**Answer:** $$
I=\left\{\begin{array}{c}
-\frac{j}{2},-4\pi \leqslant \omega \leqslant -2\pi \\
\frac{j}{2},2\pi \leqslant \omega \leqslant 4\pi \hfill \\
0, \text {otherwise} \hfill
\end{array}\right. \\
$$

# Task 3  
Compute the auto-correlation of the triangle pulse:

```python
x = np.linspace(-2, 2, 200)
y = []
for point in x:
    if (-1 <= point <= 0):
        y.append(point + 1)
    elif (0 < point <= 1):
        y.append(-point + 1)
    else:
        y.append(0)
```

```python
plt.plot(x, y) # just check that everything is ok
plt.grid()
```

```python
autocorr = np.correlate(y, y, 'full') # calculate an autocorrelation of the given function 
```

```python
print(f'length x {len(x)}')
print(f'length autocorr {len(autocorr)}')
```

```python
ac = autocorr[100:300] # cut zeros from both sides 
#(because the lentgth is 2 * len(x) - 1 and I want len(autocorr) = len(x))
```

```python
len(ac)
```

```python
ac_norm = ac / max(ac) # normlize the autocorrelation
```

### Answer

```python
fig, ax = plt.subplots(2, 1, figsize=(10,10))
ax[0].plot(x, y)
ax[0].grid(True)
ax[0].set_title('Triangle pulse')
ax[1].plot(x, ac_norm)
ax[1].grid(True)
ax[1].set_title('Auto-correlation function of the triangle pulse');
```

# Task 5.a  
The sinusoidal signal with the frequency $f = 6 kHz$ is sampled with frequency $f_{s} = 10 kHz$. Compute the apparent frequency after the signal reconstruction.


For simplicity assume that the magnitude of signal is 1, the phase is 0. Then $f(x) = sin(2\cdot6\cdot10^{3}\pi x)$

The FT of this function will give two delta functions in frequency domain located on $f = \pm 6 kHx$. Let's plot it.

```python
x = list(range(-12, 12 ,1))
y = []
for point in x:
    if point == -6:
        y.append(1)
    elif point == 6:
        y.append(1)
    else:
        y.append(0)
```

```python
# red corresponds to the sampling frequency
# blue corresponds to the initial signal frequency
# gren dash corresponds to the aliasing frequency
fig, ax = plt.subplots(figsize=(10,6))
major_ticks = list(range(-16, 16, 1))
ax.vlines(6, 0, 1)
ax.vlines(-6, 0, 1)
ax.grid()
ax.set_xticks(major_ticks);
# Then I plot our sampling frequence
ax.vlines(5, 0, 1, color = 'r')
ax.vlines(-5, 0, 1, color = 'r')
# It is obvious that frequency of the initial signal lies 
#outside the fungamental frequency range (|F|<= Fs/2) so we will achieve 
#aliasing here.
# Then I will plot the copies of initial signal near the multiplies of Fs
ax.vlines(10, 0, 1, color = 'r')
ax.vlines(-10, 0, 1, color = 'r')
ax.vlines(4, 0, 1, color = 'g', linestyle = '--')
ax.vlines(-4, 0, 1, color = 'g',linestyle = '--');
# As you can see I obtain frequency = +-4 kHz inside the fundamental frequency range
# It is the aliasing frequency
```

**Answer:** after the signal reconstruction we will get the sinusoidal signal with the frequency $f = 4kHz$ 

# Task 5.b  
The sinusoidal signal with the frequency $f = 6 kHz$ is represented by 4 signals periods sampled with the sample clock frequency $f_{s} = 15 kHz$. Are these samples enough to reconstruct the initial signal correctly.


I think that we can not reconstruct signal correctly based on these samples. Because when we sample function at frequence $f_{s}$ the sinusoidal functions yelds the same set of samples $\{sin(2\pi(f+Nf_{s})x + \varphi, N =0, \pm1, \pm2, \dots\}$ So without additional information we can't distinguish the frequency of the original function. I will try to give an example.

```python
# define the frequency of the initial signal and sampling freauency
f = 6
f_s = 15
T = 1 / f
T_s = 1 / f_s
```

```python
x = np.arange(0, 4 * T, T_s)
```

```python
y = np.sin(2 * np.pi * f * x) # initial signal
```

```python
N = 1
```

```python
y_2 = np.sin(2 * np.pi * (f + N * f_s)* x) # aliased sinusoidal signal
```

```python
plt.plot(x, y)
plt.plot(x, y_2);
plt.grid()
```

According to the graph above, both sinusoudal functions with different carrier frequencies has the same set of samples, that's why their graphs coincides.   
**Answer:** we can't reconstruct the initial signal correctly without additional information about this signal.
